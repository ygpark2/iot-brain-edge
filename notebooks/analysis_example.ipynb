# IoT-Brain-Edge Data Analysis
This notebook demonstrates how to load sensor data from ClickHouse into Spark for advanced analytics.

```python
from pyspark.sql import SparkSession

# Create SparkSession with ClickHouse JDBC/Spark Connector
spark = SparkSession.builder \
    .appName("BrainDataAnalysis") \
    .config("spark.jars.packages", "com.clickhouse.spark:clickhouse-spark-runtime-3.4_2.12:0.7.3") \
    .getOrCreate()

print("Spark Session Created.")
```

```python
# Load Raw Frames from ClickHouse
df_raw = spark.read.format("jdbc") \
    .option("url", "jdbc:ch://clickhouse:8123/brain") \
    .option("dbtable", "rawframes") \
    .option("user", "brain") \
    .option("password", "brain") \
    .option("driver", "com.clickhouse.jdbc.ClickHouseDriver") \
    .load()

df_raw.show(5)
```
